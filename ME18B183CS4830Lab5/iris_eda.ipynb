```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
sns.pairplot(iris, hue="Species", size=3)
plt.savefig('3.eps')
```

```python
iris.boxplot(by="Species", figsize=(12, 6))
plt.savefig('3.0.eps')
```
